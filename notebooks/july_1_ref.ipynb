**Why there is a bracket sometimes but sometimes not?**

bracket () is a function call. .head() means we are calling the funciton head() on the dataframe. You can put input arguments in the bracket (e.g. df.head(10) returns the first 10 rows instead of the default 5 rows.) .shape, .columns are not functions, they are attributes of the dataframe. So we don't need to use brackets () to call them.

No worries if you don't understand this now. And in case you used brackets () for attributes, it is still OKAY! You might just get a `TypeError` error message, which is a good reminder that you are calling an attribute instead of a function.

**Challenge question**
1. What is the accuracy for each subject? Who has the highest accuracy? (Hint: you can use `groupby` to group the dataframe by subject. This is very similar to the `aggregate` function in R.)
2. What is the average reaction time for each subject?

```python
# your code here
column_name = "correct"
subject_names = df["subject"].unique()
for subject in subject_names:
    # df.groupby("subject")
    subject_df = df[df["subject"] == subject]
    # for each subject's sub dataframe, calculate...
    mean_reaction_time = subject_df[column_name].mean()
    print(f"Subject: {subject}, accuracy: {mean_reaction_time}")
# df["correct"].mean()
```
